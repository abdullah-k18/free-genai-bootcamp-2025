```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from crewai import Agent, Task, Crew
```

```python
from google.colab import userdata
import os

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
os.environ["OPENAI_MODEL_NAME"] = 'gpt-4o-mini'
```

```python
sentence_translator = Agent(
    role="Sentence Translator",
    goal="Translate English {sentence} into Urdu in a structured manner.",
    backstory=(
        "You help students learn Urdu by translating English sentences."
        "Instead of directly giving the translated sentence, you provide a structured breakdown."
        "Your output includes a table with English words, their Urdu translations, word type, and pronunciation."
        "Additionally, you explain Urdu sentence structure for the given translation."
    ),
    allow_delegation=False,
    verbose=True,
)

translation_checker = Agent(
    role="Translation Checker",
    goal="Validate and provide feedback on student-translated sentences.",
    backstory=(
        "You evaluate student-translated Urdu sentences. If correct, you encourage them to try another."
        "If incorrect, you do not reveal the correct translation but provide hints and guidance to help them improve."
    ),
    allow_delegation=False,
    verbose=True,
)

clue_giver = Agent(
    role="Clue Giver",
    goal="Provide hints and guidance for Urdu sentence structure.",
    backstory=(
        "You assist students by offering clues about sentence structure, helping verbs, and grammar nuances."
        "You ensure students learn by providing hints without directly giving the answers."
    ),
    allow_delegation=False,
    verbose=True,
)

translate_task = Task(
    description=(
        "1. Analyze the given English sentence and identify key elements.\n"
        "2. Translate the sentence into Urdu without revealing the full translation immediately.\n"
        "3. Present a structured table with word-wise breakdown including English words, their Urdu translations, types, and pronunciation.\n"
        "4. Explain the Urdu sentence structure for the provided sentence.\n"
        "5. Highlight any important differences in word order, helping verbs, and grammar."
    ),
    expected_output="A structured table with English-to-Urdu translation, word types, and pronunciation.",
    agent=sentence_translator,
)

check_task = Task(
    description=(
        "1. Review the student's Urdu sentence.\n"
        "2. If correct, acknowledge and encourage another attempt.\n"
        "3. If incorrect, do NOT reveal the correct sentence but offer additional clues.\n"
        "4. Focus on improving the student's understanding rather than giving direct answers."
    ),
    expected_output="Feedback on the student's translation with additional clues if needed.",
    agent=translation_checker,
)

clue_task = Task(
    description=(
        "1. Provide guidance on Urdu sentence structure, helping verbs, and word order.\n"
        "2. Offer hints on forming correct Urdu sentences without giving direct answers.\n"
        "3. Ensure students understand concepts through progressive hints rather than direct solutions."
    ),
    expected_output="Hints and guidance for improving sentence formation in Urdu.",
    agent=clue_giver,
)

urdu_learning_crew = Crew(
    agents=[sentence_translator, translation_checker, clue_giver],
    tasks=[translate_task, check_task, clue_task],
    verbose=2
)

result = urdu_learning_crew.kickoff(inputs={"sentence": "Where are you going?"})
print(result)
```
